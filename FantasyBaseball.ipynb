```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import sqlite3

# Connecting to SQLite Database
conn = sqlite3.connect('Sqlite/FantasyBaseball.sqlite')
```

```python
# Load Batter Data
batters2019_data_to_load = "Data Sets/mlb-player-stats-Batters (2019).csv"
batters2018_data_to_load = "Data Sets/mlb-player-stats-Batters (2018).csv"
batters2017_data_to_load = "Data Sets/mlb-player-stats-Batters (2017).csv"
batters2016_data_to_load = "Data Sets/mlb-player-stats-Batters (2016).csv"
batters2015_data_to_load = "Data Sets/mlb-player-stats-Batters (2015).csv"
batters2014_data_to_load = "Data Sets/mlb-player-stats-Batters (2014).csv"
batters2013_data_to_load = "Data Sets/mlb-player-stats-Batters (2013).csv"
batters2012_data_to_load = "Data Sets/mlb-player-stats-Batters (2012).csv"
batters2011_data_to_load = "Data Sets/mlb-player-stats-Batters (2011).csv"
batters2010_data_to_load = "Data Sets/mlb-player-stats-Batters (2010).csv"
```

```python
# Read batters Data File and store into Pandas DataFrames
batters2019_data = pd.read_csv(batters2019_data_to_load)
batters2018_data = pd.read_csv(batters2018_data_to_load)
batters2017_data = pd.read_csv(batters2017_data_to_load)
batters2016_data = pd.read_csv(batters2016_data_to_load)
batters2015_data = pd.read_csv(batters2015_data_to_load)
batters2014_data = pd.read_csv(batters2014_data_to_load)
batters2013_data = pd.read_csv(batters2013_data_to_load)
batters2012_data = pd.read_csv(batters2012_data_to_load)
batters2011_data = pd.read_csv(batters2011_data_to_load)
batters2010_data = pd.read_csv(batters2010_data_to_load)
```

```python
#Add a column for the year of each dataset
batters2019_data['Year'] = 2019
batters2018_data['Year'] = 2018
batters2017_data['Year'] = 2017
batters2016_data['Year'] = 2016
batters2015_data['Year'] = 2015
batters2014_data['Year'] = 2014
batters2013_data['Year'] = 2013
batters2012_data['Year'] = 2012
batters2011_data['Year'] = 2011
batters2010_data['Year'] = 2010
```

```python
batters2019_data.head()
```

```python
#combine all batters datasets into one
batters = pd.concat([batters2019_data, batters2018_data, batters2017_data, batters2016_data, batters2015_data, 
           batters2014_data, batters2013_data, batters2012_data, batters2011_data, batters2010_data])
```

```python
print(len(batters))
```

```python
# Print out null values of all columns of `df`
print(batters.isnull().sum(axis=0).tolist())
```
